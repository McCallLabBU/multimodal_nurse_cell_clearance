```python
import pandas as pd
import seaborn as sns
color_scheme = ["#ffd366","#f6a437","#ec7014","#b0b0b0","#758a9b","#233c51"]
```

```python
quantification_summary = pd.read_csv("/Users/sbandya/Desktop/multimodal_nurse_cell_clearance/persisting_nuclei_quantification.tsv",sep='\t',header=0, index_col=0)
quantification_summary
print(quantification_summary.columns)
```

```python
cytoskeletal_genes = [ 'GR1 > LexARNAi', 'GR1 > cue RNAi','GR1 > TpnC41C','GR1 > Scp1 ','GR1 > Unc115b RNAi','GR1 > Tm1 RNAi','GR1 > Act79B RNAi ' ]
cytoskeletal_genes_quant = quantification_summary.loc[cytoskeletal_genes]
cytoskeletal_genes_quant_perc = cytoskeletal_genes_quant.iloc[:, :].apply(lambda x: x / x.sum(), axis=1) * 100
ax = cytoskeletal_genes_quant_perc.plot(kind="bar", stacked=True, alpha=1, width=0.5, color=color_scheme, fontsize=20,figsize=(6, 6))
xax = ax.axes.get_xaxis()
xax.set_visible(False)
ax.legend(loc='right', ncol=1, frameon=False,fontsize=20, bbox_to_anchor=(1.5,0.5))
ax.figure.savefig("/Users/sbandya/Desktop/multimodal_nurse_cell_clearance/cytoskeletal_pncn.png", bbox_inches="tight" )
```

```python

toll_genes = ['GR1 > LexARNAi',"GR1 > PGRP-SD RNAi ","GR1 > PGRP-LC RNAi","GR1 > Drs RNAi","GR1 > Drsl4 RNAi","GR1 > Listericin OE (33056)","GR1 > Listericin OE (33058)"]
toll_genes_quant = quantification_summary.loc[toll_genes]
toll_genes_quant_perc = toll_genes_quant.iloc[:, :].apply(lambda x: x / x.sum(), axis=1) * 100
ax = toll_genes_quant_perc.plot(kind="bar", stacked=True, alpha=1, width=0.5, color=color_scheme, fontsize=20,figsize=(6, 6))
xax = ax.axes.get_xaxis()
xax.set_visible(False)
ax.legend(loc='right', ncol=1, frameon=False,fontsize=20, bbox_to_anchor=(1.5,0.5))
ax.figure.savefig("/Users/sbandya/Desktop/multimodal_nurse_cell_clearance/toll_pncn.png", bbox_inches="tight" )
```
